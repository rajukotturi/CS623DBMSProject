```python
# Define connection parameters
conn_params = {
    "host": "localhost",
    "port": 5432,
    "database": "sbmsfinal",
    "user": "postgres",
    "password": "Eramanna4445@",
}
# Connect to the database
conn = psycopg2.connect(**conn_params)

# Create a cursor object to execute queries
cursor = conn.cursor()

try:
    # Disable auto-commit to enable transaction management
    conn.autocommit = False
    conn.set_isolation_level(psycopg2.extensions.ISOLATION_LEVEL_SERIALIZABLE)

    # Begin transaction
    cursor.execute("BEGIN")
    

    # Transaction 3: Update product p1  to pp1 in Stock
    update_stock_product = "UPDATE product SET prodid = %s WHERE prodid = %s"
    params = ("pp1", "p1")

    cursor.execute(update_stock_product, params)
    # when we update the product id to p1 in product table it can automatically updated in the stock table because it is
    # a child table containing foreign key constraint
    
    # Transaction 1: Deleting P1
    delete_product = "DELETE FROM Product WHERE prodid = %s"
    delete_stock = "DELETE FROM Stock WHERE prodid = %s"
    params = ("p1",)
    cursor.execute(delete_product, params)
    cursor.execute(delete_stock, params)

    # Commit the transaction
    conn.commit()

except (psycopg2.DatabaseError, psycopg2.IntegrityError) as e:
    # Roll back the transaction if any query fails
    conn.rollback()
    print("Transaction failed:", e)


# Select all rows from Product table
select_product = "SELECT * FROM product"
cursor.execute(select_product)
result_product = cursor.fetchall()

# Print the results
print("Product table after transaction:")
for row in result_product:
    print(row)

# Select all rows from Stock table
select_stock = "SELECT * FROM Stock"
cursor.execute(select_stock)
result_stock = cursor.fetchall()

# Print the results
print("\nStock table after transaction:")
for row in result_stock:
    print(row)
```
